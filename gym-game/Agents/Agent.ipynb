```python
from abc import ABC, abstractmethod

class Agent(ABC):
    
    @abstractmethod 
    def getName(self):
        pass
    
    @abstractmethod 
    def getPos(self):        
        pass
    
    @abstractmethod 
    def setPos(self, initialPosition):
        pass
    
    @abstractmethod 
    def getTime(self):
        pass
    
    @abstractmethod 
    def MoveNorth(self, m):
        pass
    
    @abstractmethod 
    def MoveSouth(self, m):
        pass
    
    @abstractmethod 
    def MoveEast(self, m):
        pass
    
    @abstractmethod 
    def MoveWest(self, m):
        pass
    
    @abstractmethod 
    def StayStill(self, m):
        pass
    
    #@abstractmethod 
    #def NextMove(self, m):
    #    pass
    
    @abstractmethod 
    def setInitPos(self, initialPosition):
        pass
    
    @abstractmethod 
    def getHistory(self):
        pass
    
    @abstractmethod 
    def stop(self):
        pass
    
    @abstractmethod 
    def getState(self):
        pass
    
  #  @abstractmethod 
  #  def selectAction(self, state):
  #      pass
    
```
