```python
import random
import nbimporter
import sys
if('..' not in sys.path):
    sys.path.insert(0,'..')
#from Maths.Action import Action
from  Maths.Cord import Cord
import collections
#sys.path.remove('../')


class Maze:
    def __init__(self,mazePattern):
        self.mazeString=mazePattern
        self.interpret()
        
    def interpret(self):
        parts=self.mazeString.split(",")
        self.name=parts[0]
        self.width=int(parts[1])
        self.height=int(parts[2])
        
        if(len(parts[7])!=self.width*self.height):
            print("Dimensions of maze not consistent")
            return

        InitXSet=parts[3].split("-")
        InitYSet=parts[4].split("-")
        
        if(len(InitXSet)!=len(InitYSet)):
            print("Entrance Dimensions of maze not consistent")
            return
        
        randStart= random.randint(0,len(InitXSet)-1)
        #print(randStart,len(InitXSet))
        self.Initial= Cord(int(InitXSet[randStart]),int(InitYSet[randStart]))
        ExitXSet=parts[5].split("-")
        ExitYSet=parts[6].split("-")
        
        if(len(ExitXSet)!=len(ExitYSet)):
            print("Exit Dimensions of maze not consistent")
            return
        
        self.Exits=[]
        for k in range(0,len(ExitXSet)):
            Exit= Cord(int(ExitXSet[k]),int(ExitYSet[k]))
            self.Exits.append(Exit)
        

        self.maze= [[] for t in range(self.height)]
        MazeCells=list(parts[7])
        for i in range(0, self.height):
            row=""
            rowC=""
            for j in range(0,self.width):
                b=int(MazeCells[j*self.width+i])
                self.maze[i].insert(j,b)
                row+=str(b)
                rowC+=Cord(i,j).CordToString()
            #print(row, i, rowC)
        
                            
        
        if(self.maze[self.Initial.X][self.Initial.Y]!=2):            
            print("Entrances not consistent")
            return
        
        for Exit in self.Exits:
            if(self.maze[Exit.X][Exit.Y]!=3):
                print("Exits not consistent")
                return                   
        
       # self.printMaze()
        
    def printMaze(self):
        print("Name: "+self.name)
        print(str(self.height)+"x"+str(self.width))
        print("Start: ("+str(self.Initial.X)+", "+str(self.Initial.Y)+")")
        exits=""
        for exit in self.Exits :
            exits+="("+str(exit.X)+", "+str(exit.Y)+") "
        
        print("End: "+exits)
        for i2 in range(self.width-1,-1,-1) :
            row=""
            for j2 in range(0,self.height) :
                row+=str(self.maze[j2][i2])
                #print(j2,i2,self.maze[j2][i2])
            
            print(row, i2)
      
    def CheckWall(self, c) :
        result=False 
        if (self.maze[c.X][c.Y]==1 or self.maze[c.X][c.Y]==5):
            result=True 
        return result 
    
    
    def CheckClear(self, c) :
        result=False 
        if(c.X<0 or c.X>=self.height or c.Y<0 or c.Y>=self.width): 
            return False 
        if (self.maze[c.X][c.Y]==0 or self.maze[c.X][c.Y]==2 or self.maze[c.X][c.Y]==3):
            result=True 
        return result 
    
    def returnAllClear(self):
        clear=[]
        for i in range(self.height):
            for j in range(self.width): 
                if (self.maze[i][j]==0):
                    clear.append(Cord(i,j))
        return clear
    
    def returnAllClearString(self):
        clear=""
        count=0
        for i in range(self.height):
            for j in range(self.width): 
                if (self.maze[i][j]==0):
                    clear+=Cord(i,j).CordToString()
                    count+=1
        if(count==0):
            return "X"
        return clear
    
    def returnMoving(self):
        moving=""
        for i in range(self.height):
            for j in range(self.width): 
                if (self.maze[i][j]==5):
                    moving+=Cord(i,j).CordToString()
        return moving
    
    def CheckEntrance(self, c) :
        result=False 
        if (self.maze[c.X][c.Y]==2): 
            result=True 
        return result 
    
        
    def CheckExit(self, c) :
        result=False 
        if (self.maze[c.X][c.Y]==3):
            result=True 
        return result 
    
    def WhichWayIsClear(self, c, exit_check=False) :
        directions= [] 
        if(exit_check):
            for exit in self.Exits:
                if (c.equals(exit)):
                    directions.append(4)
                    return directions
        if(c.X<0 or c.X>=self.height or c.Y<0 or c.Y>=self.width) :
            return [] 
        
        if(c.X==0) :
            if (self.CheckClear( Cord(c.X,c.Y+1))):
                directions.append(0) 
        elif (c.X==self.height-1) :
            if (self.CheckClear( Cord(c.X,c.Y-1))):
                directions.append(1) 
        else :
            if (self.CheckClear( Cord(c.X,c.Y-1))):
                directions.append(1) 
            if (self.CheckClear( Cord(c.X,c.Y+1))):
                directions.append(0) 
            
        
        if(c.Y==0) :
            if (self.CheckClear( Cord(c.X+1,c.Y))):
                directions.append(2) 
        elif (c.Y==self.width-1) :
            if (self.CheckClear( Cord(c.X-1,c.Y))):
                directions.append(3) 
        else :
            if (self.CheckClear( Cord(c.X-1,c.Y))):
                directions.append(3) 
            if (self.CheckClear( Cord(c.X+1,c.Y))):
                directions.append(2) 
            
        directions.append(4) 
        return directions
    
    def getInitialX(self) :
        return self.Initial.X
    
    def getInitialY(self) :
        return self.Initial.Y
    
    def getHeight(self) :
        return self.height
    
    def getWidth(self) :
        return self.width
    
    def getExit(self) :
        return self.Exits
    
    
    def getMazePart(self,x, y) :
        if(x<0 or y<0 or x>=self.width or y>=self.height) :
            return 1
        
        return self.maze[x][y]

    def printRoute(self, route):
        for i in range(self.width):
            row=""
            for j in range(self.height):
                c= Cord(i,j)                
                row+= "X" if (c in route) else str(self.maze[i][j])
            
            print(row)
        
        
    
    def getNeighbours(self, c):
        neighbours= []
        s= Cord(c.X+1,c.Y)
        n= Cord(c.X-1,c.Y)
        e= Cord(c.X,c.Y+1)
        w= Cord(c.X,c.Y-1)
        if(c.X==0):
            if (self.CheckClear(s)):
                neighbours.append(s)
        elif (c.X==self.height-1):
            if (self.CheckClear(n)):
                 neighbours.append(n)
        else:            
            if (self.CheckClear(s)):
                 neighbours.append(s)
            if (self.CheckClear(n)):
                  neighbours.append(n)
            
        
        if(c.Y==0):
            if (self.CheckClear(e)):
                   neighbours.append(e)
        elif (c.Y==self.width-1):
            if (self.CheckClear(w)):
                   neighbours.append(w)
        else:            
            if (self.CheckClear(e)):
                   neighbours.append(e)
            if (self.CheckClear(w)):
                   neighbours.append(w)
            
        
        return neighbours
    

    def getValue(self, r, c):
        if((r<self.height) and (r>=0) and (c<self.width) and (c>=0)):
            return self.maze[r][c]
        return -1
    
    def WallMove(self, blocked):
        #print("Wall Move")
        blocked.append(self.Initial)
        for exit in self.Exits:
            blocked.append(exit)
            
        for i in range(self.height):
            for j in range(self.width): 
                if (self.maze[i][j]==5):
                    possibleDirections=self.WhichWayIsClear(Cord(i,j))
                    direction=possibleDirections[random.randint(0,len(possibleDirections)-1)]
                    
                    nextCord=self.getNextCord(Cord(i,j),direction)
                    
                    while(nextCord in blocked):
                        direction=possibleDirections[random.randint(0,len(possibleDirections)-1)]
                        nextCord=self.getNextCord(Cord(i,j),direction)
                        
                    #print(Cord(i,j).CordToString(), direction, possibleDirections)
                    if(direction==0):
                        #print("North")
                        self.maze[i][j]=0
                        self.maze[i][j+1]=-1
                    elif(direction==1):
                        #print("South")
                        self.maze[i][j]=0
                        self.maze[i][j-1]=-1
                    elif(direction==2):
                        #print("East")
                        self.maze[i][j]=0
                        self.maze[i+1][j]=-1
                    elif(direction==3):
                        #print("West")
                        self.maze[i][j]=0
                        self.maze[i-1][j]=-1
        for i2 in range(self.height):
            for j2 in range(self.width): 
                if (self.maze[i2][j2]==-1):
                    self.maze[i2][j2]=5
                        
                    
    def getNextCord(self, c, d):
        cNext=Cord(c.X,c.Y)
        if(d==0):
            cNext=Cord(c.X,c.Y+1)
        elif(d==1):
            cNext=Cord(c.X,c.Y-1)
        elif(d==2):
            cNext=Cord(c.X+1,c.Y)
        elif(d==3):
            cNext=Cord(c.X-1,c.Y)
        return cNext


    def GetOptimalRoute(self):
        self.OptimalRoutes=[]
        for ChosenExit in self.Exits:
            OptimalRoute=[]
            visited = {}
            path = []  
            for i in range(0,self.height):
                for j in range(0,self.width):
                    if (self.maze[i][j]==0 or self.maze[i][j]==2 or self.maze[i][j]==3):                        
                        cell= Cord(i,j)                        
                        visited[cell]=False                    
            
            toExplore= collections.deque([])           
            toExplore.append(self.Initial)
            visited[self.Initial]= True    
            initialPath = []   
            initialPath.append(self.Initial)
            path.append(initialPath)            
  
            while len(toExplore)!=0:
                ps=""
                for p in path:
                    ps +=" or "
                    for cp in p:
                        ps+=cp.CordToString()
                
                
                v=toExplore.popleft()
                pathSoFar = path[0]

                path.remove(pathSoFar)
                
                if(v.__eq__(ChosenExit)):
                    OptimalRoute=pathSoFar
                    break
                
                for w in self.getNeighbours(v):
                    pathForNeighbour = pathSoFar.copy() 
                    if(visited[w]):
                        continue                 
                    
                    toExplore.append(w)
                    visited[w]=True     
                    pathForNeighbour.append(w)
                    path.append(pathForNeighbour)
                    
                    pathString=""
                    for vertex in pathSoFar:
                        pathString+=vertex.CordToString()
                    
                 
            
            self.OptimalRoutes.append(OptimalRoute)
            #self.printRoute(OptimalRoute)
        return self.OptimalRoutes
    
```
